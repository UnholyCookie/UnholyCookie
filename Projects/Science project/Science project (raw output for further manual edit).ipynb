```python
import numpy as np
import pandas as pd
```

```python
# List of Excel file names
files = ['6-03-23.xlsx', '9-03-23.xlsx', '20-03-23.xlsx', '22-03-23.xlsx', '23-03-23.xlsx',
         '28-02-23.xlsx']

# File name of the resulting file
result_file_name = 'crocs_2023_raw.xlsx'

# Read each Excel file into a separate data frame
dfs = [pd.read_excel('files/crocs/2023 data/' + file) for file in files]

# Concatenate the data frames into a single data frame
df_con = pd.concat(dfs, ignore_index=True)
```

```python
a = df_con['Unnamed: 1'][1:].dropna().str.replace('\u2028', '\n')
a = a.str.replace('\n    \n', '\n\n')
a = a.str.replace('\n      \n', '\n\n')
```

```python
male_locs = []
err = []

for i in range(len(a)):
    b = pd.Series(a.iloc[i].split('\n'))
    
#     skipping common text values
    if np.any((b == 'Areas') | (b == '—')):
        continue
#     working around typos
    try:
        try:
            c = b[np.where(b.str.find('Male') != -1)[0]].values[0]
        except:
            c = b[np.where((b.str.find('male') != -1) & (
                b.str.find('Female') == -1) & (
                b.str.find('female') == -1))[0]].values[0]
    except:
        if np.all((len(b[np.where(b.str.find('Male') != -1)[0]]) == 0) &
                    b[np.where((b.str.find('male') != -1) & (
                    b.str.find('Female') == -1) & (
                    b.str.find('female') == -1))[0]]):
            continue
        err.append(i)
        continue
#     data research logic client requirement
    if c.find('—') != -1:
        continue

    d = c[c.find('(')+1:c.find(')')].split('-')[0]

    male_locs.append(d)
```

```python
def kid_areas_calc(num_kids):
    kid_areas = []

    for i in range(len(a)):
        b = pd.Series(a.iloc[i].split('\n\n'))

        #     skipping common text values
        if np.any((b == 'Areas') | (b == '—')):
            continue

        c = b[np.where(b.str.find(num_kids + ' kid') != -1)[0]].values

        for j in range(len(c)):
            if c[j].find('—') != -1:
                continue

            for k in range(c[j].count(num_kids + ' kid')):
                kid_areas.append(c[j][:c[j].find(':')])
    return kid_areas

kid_1_areas = kid_areas_calc('1')
kid_2_areas = kid_areas_calc('2')
kid_3_areas = kid_areas_calc('3')
kid_4_areas = kid_areas_calc('4')
kid_5_areas = kid_areas_calc('5')
```

# Final operations

```python
data = [male_locs, male_moves, male_moves_2, male_env, male_areas, male_without_kids_locs,
        male_locs_3, male_locs_4,
        female_locs, female_moves, female_moves_2, female_env, female_areas, female_without_kids_locs,
        kid_1_locs, kid_1_moves, kid_1_env, kid_1_areas,
        kid_2_locs, kid_2_moves, kid_2_env, kid_2_areas,
        kid_3_locs, kid_3_moves, kid_3_env, kid_3_areas,
        kid_4_locs, kid_4_moves, kid_4_env, kid_4_areas,
        kid_5_locs, kid_5_moves, kid_5_env, kid_5_areas
        ]

cols = ['male_locs', 'male_moves', 'male_moves_2', 'male_env', 'male_areas', 'male_without_kids_locs',
        'male_locs_3', 'male_locs_4',
        'female_locs', 'female_moves', 'female_moves_2', 'female_env', 'female_areas','female_without_kids_locs',
        'kid_1_locs', 'kid_1_moves', 'kid_1_env', 'kid_1_areas',
        'kid_2_locs', 'kid_2_moves', 'kid_2_env', 'kid_2_areas',
        'kid_3_locs', 'kid_3_moves', 'kid_3_env', 'kid_3_areas',
        'kid_4_locs', 'kid_4_moves', 'kid_4_env', 'kid_4_areas',
        'kid_5_locs', 'kid_5_moves', 'kid_5_env', 'kid_5_areas'
        ]

df = pd.DataFrame(data, index=cols).T.fillna(np.nan)
```

```python
def aggregator(sex, col):
    return df.groupby(sex + '_' + col).count().sort_values(sex + '_env', ascending=False)[
        sex + '_env']

def env_aggregator(sex):
    return df.groupby(sex + '_env').count().sort_values(sex + '_locs', ascending=False)[
        sex + '_locs']
```

# Location aggregation

```python
male_locs_agg = aggregator('male', 'locs')
female_locs_agg = aggregator('female', 'locs')
kid_locs_1 = aggregator('kid_1', 'locs')
kid_locs_2 = aggregator('kid_2', 'locs')*2
kid_locs_3 = aggregator('kid_3', 'locs')*3
kid_locs_4 = aggregator('kid_4', 'locs')*4
kid_locs_5 = aggregator('kid_5', 'locs')*5
kid_locs_agg = pd.concat([kid_locs_1, kid_locs_2, kid_locs_3, kid_locs_4, kid_locs_5], 
                         axis=1).sum(axis=1).sort_values(ascending=False)
```

# Parent's locations while other parent with kids aggregation

```python
male_without_kids_locs_agg = aggregator('male', 'without_kids_locs')
female_without_kids_locs_agg = aggregator('female', 'without_kids_locs')
```

# Conditional locs aggregation

# Environment aggregation

```python
male_env_agg = env_aggregator('male')
female_env_agg = env_aggregator('female')
kid_env_1 = env_aggregator('kid_1')
kid_env_2 = env_aggregator('kid_2')*2
kid_env_3 = env_aggregator('kid_3')*3
kid_env_4 = env_aggregator('kid_4')*4
kid_env_5 = env_aggregator('kid_5')*5
kid_env_agg = pd.concat([kid_env_1, kid_env_2, kid_env_3, kid_env_4, kid_env_5], 
                         axis=1).sum(axis=1).sort_values(ascending=False)
```

# Areas aggregation

```python
male_areas_agg = aggregator('male', 'areas')
female_areas_agg = aggregator('female', 'areas')
kid_areas_1 = aggregator('kid_1', 'areas')
kid_areas_2 = aggregator('kid_2', 'areas')*2
kid_areas_3 = aggregator('kid_3', 'areas')*3
kid_areas_4 = aggregator('kid_4', 'areas')*4
kid_areas_5 = aggregator('kid_5', 'areas')*5
kid_areas_agg = pd.concat([kid_areas_1, kid_areas_2, kid_areas_3, kid_areas_4, kid_areas_5], 
                         axis=1).sum(axis=1).sort_values(ascending=False)
```
